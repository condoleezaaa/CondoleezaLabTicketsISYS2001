## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
[Help me write a python code for a budget tracker that:
 - Asks the user to input their monthly income (as a number).
 - Allows the user to input up to 3 separate expenses (also numbers), one at a time.
 - Calculates the remaining budget by subtracting total expenses from income.
 - Shows a clear warning message if the remaining budget is negative.
 - Ensures all prompts and outputs are easy to understand and user-friendly.]
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Budget Tracker Program

# Ask user for monthly income
income = float(input("Enter your monthly income: "))

# Ask user for up to 3 expenses
expenses = []
for i in range(1, 4):
    expense = float(input(f"Enter expense {i} (or 0 if none): "))
    expenses.append(expense)

# Calculate remaining budget
remaining_budget = income - sum(expenses)

# Show remaining budget
print("\nRemaining budget:", remaining_budget)

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: Your budget is negative!")
```

### 🐛 Debugging and Refinement

```python
# Document any issues and fixes
issues_found = [
    "Issue 1: User could enter a negative expense by mistake.",
    "Fix 1: Added a check to ensure expenses are not negative.",

    "Issue 2: Remaining budget was hard to read with many decimals.",
    "Fix 2: Formatted remaining budget to 2 decimal places."
]

print("🔧 DEBUGGING NOTES:")
for item in issues_found:
    print(f"- {item}")
```

```python
# REFINED VERSION - Your Working Budget Tracker
# [Paste your corrected/improved code here]
print("\n📊 Refined Budget Tracker")

# Ask user for monthly income
income = float(input("Enter your monthly income: "))

# Ask user for up to 3 expenses, ensure they are not negative
expenses = []
for i in range(1, 4):
    while True:
        expense = float(input(f"Enter expense {i} (or 0 if none): "))
        if expense < 0:
            print("❌ Expense cannot be negative. Please enter a valid amount.")
        else:
            expenses.append(expense)
            break

# Calculate remaining budget
remaining_budget = income - sum(expenses)

# Show remaining budget with 2 decimal places
print(f"\nRemaining budget: ${remaining_budget:.2f}")

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: Your budget is negative!")
```
