```python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")

expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Calculate percentage of income used
percentage_income = (expense_amount / monthly_income) * 100

# Expense categorisation logic
if expense_amount >= 100:
    category = "Major Expense"
elif expense_amount >= 50:
    category = "Moderate Expense"
elif expense_amount >= 10:
    category = "Minor Expense"
else:
    category = "Small Purchase"

# Feedback for the user
print("\n--- Expense Summary ---")
print("Description:", expense_description)
print("Amount: $", round(expense_amount,2))
print("Category:", category)

print("This expense used", round(percentage_income,2), "% of your monthly income.")

# Advice message
if percentage_income >= 50:
    print("Warning: This is a very large part of your income.")
elif percentage_income >= 25:
    print("This is a significant expense. Plan carefully.")
elif percentage_income >= 10:
    print("This is a moderate expense.")
else:
    print("This is a small purchase and within budget.")
```
